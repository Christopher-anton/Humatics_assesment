```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
dat=pd.read_csv('data.csv')
```

```python
fig1 = px.line(dat, x = 'time [s]', y = 'velocity [m/s]', title='Time [s]  vs Velocity [m/s]')
fig1.show()
```

```python
new_v = dat['velocity [m/s]'].interpolate(method='polynomial', order=2)
```

```python
fig2 = px.line(dat, x = 'time [s]', y = new_v, title='Time [s]  vs Velocity [m/s](interpolated)')
fig2.show()
```

```python
time = dat['time [s]']
```

```python
acc = np.diff(new_v)/np.diff(time) 
acc =np.append(acc,acc[-1])
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(time,new_v, label = "Velocity [m/s]")
ax.plot(time,acc, label ="Acceleration [m/s^2]" )
ax.set_xlabel('Time [s]')
ax.set_title('Time vs Velocity & Acceleration')
plt.legend()
plt.show()
```
